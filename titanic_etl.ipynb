```python
import pandas as pd
import numpy as np
import sqlite3
```

```python
df = pd.read_csv('dataset/train.csv')
df
#loading in the csv file into my system, seeing the initial length before dropping initial values
```

```python
column_names = df.columns.tolist()
column_names
#getting a list of column names so I know which one to get rid of
```

```python
del df['Cabin']
#note: I was hesitant to get rid of Cabin, but decided to delete it because nearly 80% of it was empty values
```

```python
df
#making sure that the values are correct and Cabin is gone
```

```python
median_value = df['Age'].median()
#to fill in the empty values, I chose to fill in all the 'NaN' values with the median 
df['Age'] = df['Age'].fillna(median_value)
df
```

```python
df['Embarked'] = df['Embarked'].fillna("S")
#making sure to filling any remaining values
```

```python
df['Sex'] = df['Sex'].astype('category')
```

```python
names = df['Name']

df['Title'] = df['Name'].str.extract(r'([A-Za-z]+)\.')
df
#creating a new category for the title, which will help us for finally making our analysis
```

```python
conn = sqlite3.connect("titanic.db")
result = pd.read_sql_query("SELECT COUNT(*) FROM passengers", conn)
result
```

```python
result = pd.read_sql_query("""
    SELECT Pclass, AVG(Survived) as survival_rate
    FROM passengers
    GROUP BY Pclass
""", conn)
result
```

```python
result = pd.read_sql_query("""
    SELECT Title, AVG(Survived) as survival_rate, COUNT(*) as count
    FROM passengers
    GROUP BY Title
    ORDER BY count DESC
""", conn)
result
```

This ETL pipeline cleans the Titanic passenger dataset and loads it into SQLite for analysis. I dropped the Cabin column after running early tests because 77% of its values were missing, making it unreliable and impossible to fill in a meaningful way. I replaced missing Age values with the median age, since I felt a value in line with the existing trend in the data would introduce less distortion than the mean, which is more sensitive to outliers. I extracted a Title column from the Name field (Mr/Mrs/Miss/Master/etc.) because titles capture demographic and social information that Sex and Age alone don't fully represent (for example, distinguishing young boys ("Master") from adult men ("Mr"), or unmarried women ("Miss") from married women ("Mrs")).

My three verification queries show: 
(1) a row count of 891, confirming no data was lost or duplicated during the load step; 
(2) survival rate by Pclass, which shows a clear downward trend as class decreases (1st class ~63%, 2nd class ~47%, 3rd class ~24%)
(3) survival rate by Title, which shows that despite adult men ("Mr") making up the largest group of passengers (517 of 891), they had by far the lowest survival rate (~16%)

While women and young boys survived at much higher rates, reflecting the "women and children first" evacuation pattern by outcome, they did not make up the passenger composition.
